```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
 movie_data = pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape 
```

```python
Selected_Features =['genres','keywords','tagline','cast','director']
print (Selected_Features)
```

```python
# replace null values 
for feature in Selected_Features:
  movie_data[feature] = movie_data[feature].fillna('')
```

```python
#combining features 
combined_features = movie_data['genres'] +''+movie_data['keywords']+''+movie_data['tagline']+''+movie_data['cast']+''+movie_data['director']
print(combined_features)
```

```python
#combining text data in num value using vecteriser 
vecteriser = TfidfVectorizer()
```

```python
features_vector = vecteriser.fit_transform(combined_features)
```

```python
print(features_vector)
```

```python
#sililarity score 
similarity = cosine_similarity(features_vector)
```

```python
print(similarity)
```
